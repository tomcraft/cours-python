```
#importer vos librairies
import pandas as pd
import sklearn as sk
from sklearn import model_selection
from sklearn import linear_model
```

```
#charger les données
price = pd.read_csv("price_availability.csv", sep=";")
listing = pd.read_csv("listings_final.csv", sep=";")
```

```
#cleanup
listing.drop("Unnamed: 0", axis="columns", inplace=True)
```

```
price.head()
```

```
listing.head()
```

```
#vérifier si tous les individus ont bien un prix
price.isna()['local_price'].sum()
```

##Données d'entrée

```
#définir 2 variables de travail
#X := les features à utiliser = (presque) tout le reste
#Y := la target (prix) = local_price
```

```
listing_price_mean = price.groupby("listing_id")['local_price'].mean()
```

```
listingPrice = listing.merge(listing_price_mean, on="listing_id")
listingPriceX = listingPrice[["person_capacity", "beds", "bedrooms", "bathrooms"]]
listingPriceY = listingPrice[["local_price"]]
```

```
listingPrice.head()
```

```
#utiliser la méthode split de sklearn en splitant avec un alpha=30 et un random state=42 
#zafficher la shape de vos données

X_train, X_test, y_train, y_test = model_selection.train_test_split(listingPriceX, listingPriceY, train_size=0.30, random_state=42)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```
#créer l'objet de régression et entrainer le sur notre ensemble d'entraînement
reg = linear_model.LinearRegression().fit(X_train, y_train)
```

```
#afficher les coefficients
#que remarquez vous ?
print("beta_j", reg.coef_)
print("beta_0", reg.intercept_)
```

```
reg.predict(X_test)
```
